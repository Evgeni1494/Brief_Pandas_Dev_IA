# Étape n°2 : Nettoyage et préparation des données

## 1


 
Nettoyer les titres de colonnes :

 1.1 Supprimer les espaces au début et à la fin des titres (s’il y en a)

 1.2 Remplacer les espaces par des underscores (ceux se situant entre les mots)

 1.3 Tout mettre en minuscule


```python
# 1.1

df_pop.columns = df_pop.columns.str.strip()
df_anim.columns = df_anim.columns.str.strip()
df_veg.columns = df_veg.columns.str.strip()
```

```python
# 1.2

df_pop.columns = df_pop.columns.str.replace(' ', '_')
df_anim.columns = df_anim.columns.str.replace(' ', '_')
df_veg.columns = df_veg.columns.str.replace(' ', '_')
```

```python
# 1.3

df_pop.columns = df_pop.columns.str.lower()
df_anim.columns = df_anim.columns.str.lower()
df_veg.columns = df_veg.columns.str.lower()
```

 ## 4. 
 Pour les datasets df_anim et df_veg, ajoutez une colonne ‘type’ qui prendra

    respectivement une valeur ‘animal’ et ‘vegetal’. Une fois cette étape effectuée, regroupez

    les deux jeux de données en 1 et appelez ce DataFrame product. Attention à bien

    comprendre la structure des données pour utiliser la bonne méthode

```python
# 4 

type = ['animale'] * len(df_anim.index)
type2 = ['vegetal'] * len(df_veg.index)
df_anim = df_anim.assign(type=type)
df_veg = df_veg.assign(type=type2)

product = pd.concat([df_anim, df_veg])
```

```python
# 4
product.head()
```

## 5. 
Transformez df_pop afin de ne garder que le code du pays, le pays, l’année et la population. 

Renommer la colonne ‘value’ en ‘pop_1000_hab’.

```python
df_pop.head(2)
```

```python
# code_zone_(m49) => code_du_pays
# zone => pays
# année => année
# valeur => population

# df_pop = df_pop['code_zone_(m49)'],['zone'],['année'],['population']


df_pop = df_pop.loc[:,['code_zone_(m49)', 'zone','année','valeur']]
df_pop = df_pop.rename(columns={'code_zone_(m49)': 'code_du_pays', 'zone': 'pays','valeur':'pop_1000_hab'})
```

```python
df_pop.head()
```

 ## 6.
Transformez products afin de ne garder que les colonnes :

 area_code_(fao), area, element, item, year, type, unit, value.

```python
#  area_code_(fao), area, element, item, year, type, unit, value.

# area_code_(fao) => code_zone_(m49)
# area => zone
# element => élément
# item => produit
# year => année
# type => type
# unit => unité
# value => valeur

product = product.loc[:,['code_zone_(m49)', 'zone','élément','produit','année','type','unité','valeur']]
product = product.rename(columns={'code_zone_(m49)': 'code_du_pays', 'zone': 'pays','élément':'element','unité':'unit'})
```

```python
product.head()
```

## 7

Fusionnez df_pop avec products et nommez ce DataFrame df. 

Afin de fusionner ces jeux de données vous devez identifier les clés primaires. 

Renommer les colonnes comme sur la capture d’écran ci-dessous.

```python
df_pop.head(2)
```

```python
df = pd.merge(df_pop, product, on=['code_du_pays', 'pays', 'année'])
```

```python

list_name = ['eastern_africa','middle_africa','northern_africa','southern_africa','western_africa','northern_america','central_america',

'caribbean','south_america','central_asia','eastern_asia','southern_asia','south_eastern_asia','western_asia', 

'eastern_europe','northern_europe','southern_europe','western_europe','australia_new_zealand','melanesia','micronesia', 

'polynesia']

def preproc(list_name):
    dataframes = []
    for region in list_name:
        file_path = os.path.join('Df_regions', f'{region}.csv')
        df = pd.read_csv(file_path, usecols=['Area Code (M49)'])
        df['zone'] = region
        dataframes.append(df)
    result = pd.concat(dataframes)
    return result

df_regions = preproc(list_name)

```
