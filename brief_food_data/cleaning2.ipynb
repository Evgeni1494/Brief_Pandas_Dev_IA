```python

list_name = ['eastern_africa','middle_africa','northern_africa','southern_africa','western_africa','northern_america','central_america',

'caribbean','south_america','central_asia','eastern_asia','southern_asia','south_eastern_asia','western_asia', 

'eastern_europe','northern_europe','southern_europe','western_europe','australia_new_zealand','melanesia','micronesia', 

'polynesia']

def preproc(list_name):
    dataframes = []
    for region in list_name:
        file_path = os.path.join('Df_regions', f'{region}.csv')
        df = pd.read_csv(file_path, usecols=['Area Code (M49)'])
        df['zone'] = region
        dataframes.append(df)
    result = pd.concat(dataframes)
    return result

df_regions = preproc(list_name)

```

# 7


Créez une nouvelle colonne taux_croissance_pop_18_19 avec le taux de variation de la population entre 2018 et 2019 dans chaque pays. 

Affichez les 5 pays avec le taux de croissance démographique le plus élevé.

```python

df_2018 = pivot_df[pivot_df['year'] == 2018]
df_2019 = pivot_df[pivot_df['year'] == 2019]


df_2018_2019 = pd.merge(df_2018, df_2019, on='area', suffixes=('_2018', '_2019'))


df_2018_2019['taux_croissance_pop_18_19'] = (df_2018_2019['pop_1000_hab_2019'] - df_2018_2019['pop_1000_hab_2018']) / df_2018_2019['pop_1000_hab_2018'] * 100


taux_croissance_pop_18_19 = df_2018_2019[['area', 'taux_croissance_pop_18_19']]


taux_croissance_pop_18_19 = taux_croissance_pop_18_19.groupby(by="area").mean()


pivot_df = pd.merge(pivot_df, taux_croissance_pop_18_19, on='area', how='left')

```

# 15

Avec la disponibilité alimentaire combien d’être humains pourrait-on nourrir ? 

et avec la disponnibilité alimentaire en végétaux ?

Pour répondre à cette question on prendra la valeur de 2250 kcal pour la quantité necessaire a nourrir un etre humain.

Interpréter vos résultats. Exprimer vos résultats sous la forme de % de la population mondiale.

```python
q_15 = pivot_df[['zone','pop_1000_hab','Food supply (kcal/capita/day)','type']].copy()


total_kcal = q_15['Food supply (kcal/capita/day)'] * q_15['pop_1000_hab'] * 365
total_people = total_kcal / 2250  
total_percent = total_people.sum() / 7.9e9 * 100  


veg_kcal = q_15.loc[q_15['type'] == 'vegetal', 'Food supply (kcal/capita/day)'] * q_15['pop_1000_hab'] * 365
veg_people = veg_kcal / 2250  
veg_percent = veg_people.sum() / 7.9e9 * 100 

print(f"Avec la disponibilité alimentaire totale, on peut nourrir environ {total_percent:.2f}% de la population mondiale.")
print(f"Avec la disponibilité alimentaire en végétaux, on peut nourrir environ {veg_percent:.2f}% de la population mondiale.")
```
